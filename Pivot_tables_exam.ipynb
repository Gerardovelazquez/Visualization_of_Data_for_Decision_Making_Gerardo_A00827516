```python
#Importar la libreria para que pueda abrir el archivo excel y tablas
import pandas as pd
import numpy as np
```

```python
#Verificamos si cargo la libreria de antropometria 
%ls datasets/
```

```python
#Abrimos el archivo 1 de antropometria
data_frame_antropometria= pd.read_csv("datasets/antropometria-dataset.csv")
data_frame_antropometria.head(32)
```

```python
#Q1 Peso Promedio Sinaloa
#Como en la tabla anterior vimos que desc_ent son los estados la usaremos como index y el value sera el peso, pues es el proedio que nos solicita.
pivot_question_n1 = pd.pivot_table(data_frame_antropometria,
                       index= "desc_ent",
                       values= "peso",
                      )
pivot_question_n1
```

```python
#Q2  Median sexo 2 Yucatan

#Creamo otro pivot table  para sacar la median, usamos de nuevo de index desc_ent para tener todos los estados enlistados, ponemos el peso que es
# lo que necesitamos y le agregamos el aggfunc para que saque la median y especificamos que sea de las columnas sexo.

pivot_question_n2 = pd.pivot_table(data_frame_antropometria,
                         index="desc_ent",
                         values="peso",
                         aggfunc="median",
                         columns="sexo"
                        )
pivot_question_n2
```

```python
#Q3 valor promedio maximo de peso del sexo 1
#Repetimos los pasos de la pregunta anterior solo que en el aggfunc ponemos mean, pues aui nos solicita el promedio.
#Busco el promedio mas alto el mas grande

pivot_question_n3= pd.pivot_table(data_frame_antropometria,
                        index="desc_ent",
                        values="peso",
                        aggfunc="mean",
                        columns="sexo"
                       )
pivot_question_n3
```

```python
#Leemos el archivo
data_frame_adult=pd.read_csv("datasets/adult.csv")
data_frame_adult.head()
```

```python
#Q4 Promedio de Age Nevermarried Bachelors
#Creamos otra pivot table con los valores  solicitados, poniendo en aggfunc mean, pues se ocupa el promedio.
#Nuestro index es el marital status pq necesitaamos encontrar el nevermarried, value es age pues necesitamos la edad
# y usamos la column educaation para encontrar los bachelors

pivot_question_n4= pd.pivot_table(data_frame_adult,
                        index="marital-status",
                        values="age",
                        aggfunc="mean",
                        columns="education"
                       )
pivot_question_n4
```

```python
#Q5 Median age Male Married civ education 12..
#Repetimos los pasos anteriores solo que le cambiamos el value a hours per week, la aggfunc a median,
# y las columns la dividimos en dos para tener el peso exacto del hombre

pivot_question_n5= pd.pivot_table(data_frame_adult,
                        index="marital-status",
                        values="hours-per-week",
                        aggfunc="median",
                        columns=("education", "sex"),
                       )
pivot_question_n5 
```
