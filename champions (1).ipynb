```python
#Solving Q3 3.-How many goals did each local and away team score?

#Goles de equipos jugando de local
pivot=pd.pivot_table(df,
                     index= "team_name_home",
                     values= ["team_home_score"],
                     aggfunc= "sum",
                     columns= ["stage"],
                    )
pivot
                   
```
